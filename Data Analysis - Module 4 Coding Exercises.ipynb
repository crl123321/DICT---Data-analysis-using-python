```python
import pandas as pd
```

```python
bestsellers = pd.read_csv("bestsellers.csv")
bestsellers.head()
```

```python
bestsellers.info()
bestsellers.isnull().sum()
```

```python
bestsellers[bestsellers["Price"].isna()]
```

```python
bestsellers = bestsellers.dropna()
bestsellers.info()
bestsellers.isnull().sum()
```

```python
bestsellers.head()
```

```python
del bestsellers["Type"]
bestsellers.head()
```

```python
bsellers = bestsellers.groupby("Year", as_index = False).agg({"Reviews" : "sum"})
bsellers
```

### Are the number of reviews increasing each year?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Line plot using seaborn
sns.lineplot(x = "Year", y = "Reviews", data = bsellers, marker = "o")
plt.title("Reviews per Year")
plt.xlabel("")
plt.show()
```

```python
most = bsellers.sort_values(by = "Reviews", ascending = False)
most.head(10)

most_2 = bestsellers.groupby("Name", as_index = False).agg({"Reviews" : "sum"})
most_2_2 = most_2.sort_values(by = "Reviews", ascending = False)

most_2_2.head(10)
```

#### Q2 Which book received the most number of reviews?
##### Answer: The most reviewed book is "Where the Crawdads Sing" by Delia Owens.

```python
r_top = most_2_2[(most_2_2["Reviews"] >= 35799.0)]

sns.barplot(y = "Name", x = "Reviews", data = r_top, orient = "h", palette = "pastel")
plt.title("Top Books with most Reviews")
plt.ylabel("")
plt.xlabel("")

plt.show()
```
